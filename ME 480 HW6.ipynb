```python
# Problem 5a
# Real numbers with multiplication

# Closed:
print(type(2.54*np.pi))

# Associative:
print((2.54*np.pi)*np.sqrt(2) == 2.54*(np.pi*np.sqrt(2)))

# Identity:
print(1*2.54)

# Inverse:
print(2.54+(-2.54))

# Therefore Real numbers with multiplication is a group
```

```python
# Problem 5b
# Matrices with matrix multiplication
A = np.array([[1,0],[0,1]])
B = np.array([[-1,0],[0,-1]])

# Closed:
print(type(A*B))

# Associative:
print((A*B)*A == A*(B*A))

# Identity:
print(A*B == B)

# Inverse:
print(A + (-A))

```

```python
# Problem 7
def se2(theta,D):
    Rt = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
    d = np.array(D).T.reshape((1,2)) 
    return np.block([
                    [Rt , d.T],
                    [0,0,   1]  
                    ])

def twoDVectorMatrix(A):
    return np.array([[1,0,A[0]],
                     [0,1,A[1]],
                     [0,0,1]])

def extract2DVector(M):
    p1 = M[0:2,2]
    return p1

p = np.array([5,5])
Tp = twoDVectorMatrix(p)
Tpinv = twoDVectorMatrix(-p)
R = se2(np.radians(45), [0,0])
RM = Tp@R@Tpinv

p_start = np.array([10,10])
ps = twoDVectorMatrix(p_start)
pe = RM@ps
p_end = extract2DVector(pe)

print(p_end)
```

```python
def Rx(theta):
    c = np.cos(theta)
    s = np.sin(theta)
    rx = np.array( [ [1, 0,  0],
                   [0, c, -s],
                   [0, s,  c] ] )
    return rx

def Ry(theta):
    c = np.cos(theta)
    s = np.sin(theta)
    ry = np.array( [ [c,  0,  s],
                     [0,  1, 0],
                     [-s, 0,  c] ] )
    return ry

def Rz(theta):
    c = np.cos(theta)
    s = np.sin(theta)
    rz = np.array( [ [c, -s,  0],
                     [s,  c,  0],
                     [0,  0,  1] ] )
    return rz
```

```python
def ZXpZpp(psi,theta,phi):
    return Rz(psi)@Rx(theta)@Rz(phi)

def YXpZpp(psi,theta,phi):
    return Ry(psi)@Rx(theta)@Rz(phi)

def YXpYpp(psi,theta,phi):
    return Ry(psi)@Rx(theta)@Ry(phi)
```

```python
def Nx(n):
    x = n[0]
    y = n[1]
    z = n[2]
    nx = np.array( [ [0, -z,  y],
                     [z,  0, -x],
                     [-y,  x, 0] ] )
    return nx

def AxisAngle(n,theta):
    mag = np.linalg.norm(n)
    v = n/mag
    s = np.sin(theta)
    c = np.cos(theta)
    NM = Nx(v)
    r = np.eye(3)+s*NM+(1-c)*NM@NM
    return r
```

```python
def threeDVectorMatrix(V):
    v = np.array(V).T.reshape((1,3)) 
    return np.block([
                    [np.eye(3) , v.T],
                    [0,0,0, 1]  
                    ])

def extract3DVector(M):
    p1 = M[0:3,3]
    return p1
```

```python
def AxisAngleDistance(n,theta,D):
    R = AxisAngle(n,theta)
    d = np.array(D).T.reshape((1,3)) 
    return np.block([
                    [R , d.T],
                    [0,0,0, 1]  
                    ])

def extract3DM(M):
    p1 = M[0:3,0:3]
    return p1
```

```python
# Problem 12
def Flail(M,N,R):
    ball = sphere(r=R)
    spk = translate([0,0,R-1])(cylinder(d1=R/3 , d2=0 , h=R/1.5))
    spksum = spk
    for i in range(M):
        spknew = rotate(360/M*i, [0,1,0])(spk)
        spksum = spksum + spknew
    spkrings = spksum
    for j in range(N):
        newring = rotate(360/N*j, [0,0,1])(spksum)
        spkrings = spkrings + newring
        
    return spkrings + ball


F = Flail(6,3,10)
r.render(F)
```
